Problema 1:

$x + 8 = 15$

Planteamiento: Tenemos una ecuación de primer grado donde buscamos un número $x$ que, al sumarle 8, nos dé como resultado 15.

Despeje: Para dejar a la $x$ sola (despejarla), debemos quitar el 8. Como el 8 está sumando, pasa al otro lado de la igualdad con la operación contraria, que es la resta.$x = 15 - 8$

Resultado: Realizamos la resta simple.$x = 7$

```python
from sympy import symbols, Eq, solve

# Definir la variable simbólica 'x'
x = symbols('x')

# Definir la ecuación: x + 8 = 15
# Nota: Eq(izquierdo, derecho) representa la igualdad
ecuacion = Eq(x + 8, 15)

# Resolver la ecuación para hallar el valor de x
solucion = solve(ecuacion, x)

# Imprimir el primer elemento de la lista de soluciones
print(f"La solución para el Problema 1 es x = {solucion[0]}")
```

Problema 3:

 $3x + 4x = 21$

 Simplificación: Primero sumamos los términos semejantes. Como ambos tienen la variable $x$, sumamos sus coeficientes: $3 + 4 = 7$.Ecuación: $7x = 21$

 Despeje: El 7 está multiplicando a la $x$. Para "pasarlo" al otro lado, usamos la operación inversa (división).$x = \frac{21}{7}$

 Resultado: $x = 3$

```python
from sympy import symbols, Eq, solve

# Definir la variable
x = symbols('x')

# Definir la ecuación: 3x + 4x = 21
ecuacion = Eq(3*x + 4*x, 21)

# Resolver la ecuación
solucion = solve(ecuacion, x)

print(f"La solución para el Problema 3 es x = {solucion[0]}")
```

Problema 5:

 $6a - 2a + a = 20$

 Simplificación: Agrupamos los términos de la variable $a$. Recuerda que $a$ sola equivale a $1a$.$(6 - 2 + 1)a = 20$$5a = 20$

 Despeje: El 5 multiplica a $a$, por lo tanto, dividimos 20 entre 5.$a = \frac{20}{5}$

 Resultado: $a = 4$

```python
from sympy import symbols, Eq, solve

# Definir la variable (usamos 'a' en este caso)
a = symbols('a')

# Definir la ecuación: 6a - 2a + a = 20
ecuacion = Eq(6*a - 2*a + a, 20)

# Resolver la ecuación
solucion = solve(ecuacion, a)

print(f"La solución para el Problema 5 es a = {solucion[0]}")
```

Problema 7:

$4x + 3 = 2x + 11$

Agrupación: Pasamos todas las $x$ a un lado y los números al otro. El $2x$ pasa restando y el $3$ pasa restando.$4x - 2x = 11 - 3$

Simplificación: Realizamos las operaciones en ambos lados.$2x = 8$

Despeje: Dividimos 8 entre 2.$x = \frac{8}{2}$

Resultado: $x = 4$

```python
from sympy import symbols, Eq, solve

# Definir la variable
x = symbols('x')

# Definir la ecuación: 4x + 3 = 2x + 11
ecuacion = Eq(4*x + 3, 2*x + 11)

# Resolver la ecuación
solucion = solve(ecuacion, x)

print(f"La solución para el Problema 7 es x = {solucion[0]}")
```

Problema 9:

$9m + 4 = 5m + 20$

Transposición de términos: Agrupamos los términos con la incógnita $m$ en el lado izquierdo y los términos independientes (números) en el derecho. El $5m$ pasa restando y el $4$ pasa restando.$9m - 5m = 20 - 4$

Reducción de términos semejantes: Realizamos las restas en ambos lados de la igualdad.$4m = 16$

Despeje de la variable: El $4$ que multiplica a $m$ pasa al otro lado dividiendo al $16$.$m = \frac{16}{4}$

Resultado: $m = 4$

```python
from sympy import symbols, Eq, solve

# Definir la variable simbólica 'm'
m = symbols('m')

# Definir la ecuación: 9m + 4 = 5m + 20
ecuacion = Eq(9*m + 4, 5*m + 20)

# El comando solve encuentra el valor que satisface la igualdad
solucion = solve(ecuacion, m)

print(f"La solución para el Problema 9 es m = {solucion[0]}")
```

Problema 11:

$6a + 5 = 2a + 21$

Agrupación: Movemos las variables al lado izquierdo y los números al derecho. El $2a$ que es positivo pasa como negativo, igual que el $5$.$6a - 2a = 21 - 5$

Simplificación: Restamos los coeficientes y los valores constantes.$4a = 16$

Despeje: Dividimos el resultado constante entre el coeficiente de la variable.$a = \frac{16}{4}$

Resultado: $a = 4$

```python
from sympy import symbols, Eq, solve

# Definir la variable simbólica 'a'
a = symbols('a')

# Definir la ecuación: 6a + 5 = 2a + 21
ecuacion = Eq(6*a + 5, 2*a + 21)

# Resolver la ecuación para 'a'
solucion = solve(ecuacion, a)

print(f"La solución para el Problema 11 es a = {solucion[0]}")
```
